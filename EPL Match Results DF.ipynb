```python
import pandas as pd
import numpy as np
from numpy import random 
import matplotlib.pyplot as plt
```

```python
original_df = pd.read_csv('EPL_Set.csv', index_col='Date', parse_dates=True)
del original_df['Div']
#Adding random results with a random integer and changing all the 2's to 3's
original_df['randomResult'] = np.random.randint(0,3,len(original_df))
change_wins = original_df['randomResult'] == 2
original_df.loc[change_wins,'randomResult'] = 3
```

```python
##function to filter a team's results by w/d/l and add the points to the df
def results_and_points(df, team):
    ##Function takes a dataframe of EPL match results and a specific team and returns a dataframe containing only the rows  
    #pertaining to that team's matches and whether they resulted in a w/l/d
    is_team = (df['HomeTeam'] ==team) | (df['AwayTeam']==team)
    team_games = df.loc[is_team]
    ##boolean masks
    win = ((team_games['HomeTeam'] == team) & (team_games['FTR'] =='H')) |((team_games['AwayTeam'] ==team) & (team_games['FTR'] =='A'))
    tie = ((team_games['HomeTeam'] == team) & (team_games['FTR'] =='D')) |((team_games['AwayTeam'] ==team) & (team_games['FTR'] =='D'))
    loss = ((team_games['HomeTeam'] == team) & (team_games['FTR'] =='A')) |((team_games['AwayTeam'] ==team) & (team_games['FTR'] =='H'))
    ##applied
    team_games.loc[win,'points']=3
    team_games.loc[tie,'points']=1
    team_games.loc[loss,'points']=0
    #team_games = team_games.reset_index()
    #team_games.reindex(range(1,39))
    return team_games
```

```python
def season_and_teams(season,dataframe):
    ## Funtion that takes a season slice from original_df and applies results_and_points to every team from that season. 
    ## Returns a dictionary
    season_slice = dataframe.loc[dataframe['Season']==season]
    unique_teams = sorted(season_slice.HomeTeam.unique())
    dict_title={}
    for team in unique_teams:
        dict_title[team] = results_and_points(season_slice,team)
    return dict_title
```
